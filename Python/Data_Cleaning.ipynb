```python
reference_date = sales['order_date'].max()

reference_date
```

```python
rfm = sales.groupby('customer_id').agg({

    'order_date': lambda x: (reference_date - x.max()).days,

    'order_id': 'count',

    'revenue': 'sum'

})
```

```python
rfm.columns = [
    'Recency',
    'Frequency',
    'Monetary'
]
```

```python
rfm.head()
```

```python
rfm['R_Score'] = pd.qcut(
    rfm['Recency'],
    5,
    labels=[5,4,3,2,1]
)


rfm['F_Score'] = pd.qcut(
    rfm['Frequency'],
    5,
    labels=[1,2,3,4,5]
)


rfm['M_Score'] = pd.qcut(
    rfm['Monetary'],
    5,
    labels=[1,2,3,4,5]
)
```

```python
rfm['RFM_Score'] = (
    rfm['R_Score'].astype(int)
    +
    rfm['F_Score'].astype(int)
    +
    rfm['M_Score'].astype(int)
)
```

```python
rfm.head()
```

```python
def customer_segment(score):

    if score >= 13:
        return "Champions"

    elif score >= 10:
        return "Loyal Customers"

    elif score >= 7:
        return "Potential Loyal"

    else:
        return "At Risk"


rfm['Segment'] = rfm['RFM_Score'].apply(customer_segment)
```
